```python
# Hello!
# Input: `"/foo"` 

#     - /foo
#       - /images
#         - /foo.png  <------.
#       - /temp              | same file contents
#         - /baz             |
#           - /that.foo  <---|--.
#       - /bar.png  <--------'  |
#       - /file.tmp  <----------| same file contents
#       - /other.temp  <--------'
#       - /blah.txt -> /foo

# Output:

#       [
#          ['/foo/bar.png', '/foo/images/foo.png'],
#          ['/foo/file.tmp', '/foo/other.temp', '/foo/temp/baz/that.foo']
      ]
# map: hash => list of file names


# file_a
# file_b == file_a
# link_b
# link_c 

# [file_a, file_b, link_b]



def isDir(filename): bool

def find_files(directory, res={}, map_links):
    if not fileExists(directory):
        return
    for f in files(directory):
        if isDir(f):
            find_files(f, res)
        elif isSimLink(f):
            link = getParent(f)
            add_to_map(map_links, link, f)
        else:
                
            h5 = file.size(f)
            if h5 not in res:
                res[h5] = set([f])
            else:
                res[h5].add(f)
            if f in map_links:
                res[h5] += map_links[f]

def transform_to_list_of_dupls(map_res):
    return [list(f) for f in filter(lambda v: len(v) > 1, map_res.values())]


def find_dupls(dir):
    map_res = {}
    find_files(dir, map_res)
    res = transform_to_list_of_dupls(map_res)
    return res


def get_dupls(files):
    h = {}
    for f in files:
        h5 = hash.md5(f)
        if h5 not in h:
            h[h5] = [f]
        else:
            h[h5].append(f)
    return transform_to_list_of_dupls(h)



def find_all_dupls(dir):
    res = []
    res1 = find_dupls(dir)
    for ff in res1:
        res2 = get_dupls(ff)
        for r in res2:
            res.append(r)
    return res
    

res = find_all_dupls('/foo')


# O(n) + 
```
