```python
from functools import reduce
def x_contains_any_y(x, y):
    for y_ in y:
        if y_ in x:
            return True
    return False
xx = []
for i in range(len(arr)):
    for j in range(len(arr[i])):
        if arr[i][j] == 1:
            yy = []
            y = None
            for x in xx:
                if x_contains_any_y(x,[(i-1, j),(i+1,j),(i,j-1),(i,j+1)]):
                    x.add((i,j))
                    y = x if y is None else y.union(x)
                else:
                    yy.append(x)
            yy.append(set([(i,j)]) if y is None else y)
            xx = yy
s = reduce(lambda x, y: x if len(x)>len(y) else y, xx)
print(len(s))

for i in range(len(arr)):
    for j in range(len(arr[i])):
        print('1 ' if (i,j) in s else '0 ', end ="")
    print()
#s
```
