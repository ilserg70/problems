```python
import sys, os, re
```

```python
def get_files(root, finfo=set(), exclude=None, del_prefix="",):
    prog = re.compile(exclude) if exclude is not None else None
    for path, subdirs, files in os.walk(root):
        for name in files:
            f = os.path.join(path, name)
            if not prog.match(f):
                b = os.path.getsize(f)
                finfo.add((b, f.replace(del_prefix,"")))
                #print("{} - {}".format(b, f))
        for name in subdirs:
            get_files(os.path.join(path, name), finfo, exclude, del_prefix)
root = "/Users/silinskiy/Documents/mygithub/coursera/"
exclude = "(\.DS_Store|\.git)"
finfoset=set()
get_files(root, finfoset, exclude, root)
finfo = sorted(list(finfoset), key=lambda x: x[0], reverse=True)
total_size = sum(map(lambda x: x[0], finfo))
print("{}B or {}kB or {}mB".format(total_size, int(total_size/1024), int(total_size/(1024*1024))))
finfo
```
